```python
class Date(object):
    '''
    doc data
    '''

    def __init__(self, day=0, month=0, year=0):
        self.day = day
        self.month = month
        self.year = year

    @classmethod
    def from_string(cls, date_as_string):
        print("date_as_string", date_as_string)
        date_list = date_as_string.split('-')
        day = int(date_list[0])
        month = int(date_list[1])
        year = int(date_list[2])
        date2 = cls(day, month, year)
        return date2
```

```python
date1 = Date(25, 12, 2021)
date2 = Date.from_string("25-2-2222")
```

```python
import pandas as pd

df = pd.DataFrame()
df['nilai'] = [1, 2, 3, 4, 5, 6]
df.head(len(df))
```

```python
# jika nilai > 3 ==> merah

def is_merah(data):
    new_data = []
    for i in data:
        
        if i > 3:
            y = 'merah'
        else:
            y = 'putih'
        new_data.append(y)
    return new_data

data = df['nilai']


df['warna'] = is_merah(data)
df.head()
```

```python
class Bunga:
    def __init__(self, nama, age, asal):
        # Data        
        self.nama = nama
        self.age = age
        self.asal = asal
        
    def lihat(self):
        print(f"Ini bunga {self.nama} berumur {self.age}")

    def get_asal(self):
        print("Bunga berasal dari "+ self.asal)


class Matahari(Bunga):
    def __init__(self, nama, age, asal):
        super().__init__(nama, age, asal)

    def jumlah_biji(self):
        return "ribuan"

    
class Sepatu(Bunga):
    def __init__(self, nama, age, asal):
        super().__init__(nama, age, asal)

    def jumlah_jumlah_kelopak(self):
        return "5"
```
