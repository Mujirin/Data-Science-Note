```python
### Mapping function

def transform_value(row, mapping, column_name): # Apakah argumen "row" pad fun(row, args..) merupakan format default untuk fungsi dataframe.apply??
    # Ya, argumen pertama (row) selalu dipakai sebagai arg default untuk mengidentifikasi baris pada dataframe.
    # Fungsi ini mengambil data dari sel tertentu (baris, kolom)
    data = row[column_name] # mengambil data pada cell tertentu (baris)
    value = None
    try:
        for map_ in mapping: # map_ bernilai sesuai dengan isi sel pada mapping pada urutan indeks
            if data in map_['from']: # mengecek, apakah data ada di dalam list map_['from']
                value = map_['to'] # bila ada, value = map_['to'], lalu return
                
    except Exception as e:
            value = data
            
    if not value:
        value = data
    return value
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import typing
import dask.dataframe as dd


# noinspection PyShadowingBuiltins,PyUnusedLocal
def join(
        df1: dd.core.DataFrame,
        df2: dd.core.DataFrame,
        leftKey: list,
        rightKey: list,
        type: str = "left",
        *args, **kwargs) -> typing.Tuple[dd.core.DataFrame, None]:

    if not df1.known_divisions:
        index_name = df1.index.name if df1.index.name else "index"
        df1 = df1.reset_index().set_index(index_name, sorted=False)
    if not df2.known_divisions:
        index_name = df2.index.name if df2.index.name else "index"
        df2 = df2.reset_index().set_index(index_name, sorted=False)

    if type == "cross":
        df = df1.assign(__key__=1).merge(df2.assign(__key__=1), on="__key__", suffixes=['_x', '_y'])\
            .drop("__key__", axis=1)
    elif type == "concat-row":
        df = dd.concat([df1, df2], axis=0)
    elif type == "concat-column":
        df = dd.concat([df1, df2], axis=1)
    elif leftKey and rightKey:
        df = dd.merge(df1, df2, left_on=leftKey, right_on=rightKey, suffixes=['_x', '_y'], how=type)
    elif leftKey:
        df = dd.merge(df1, df2, on=leftKey, suffixes=['_x', '_y'], how=type)
    elif rightKey:
        df = dd.merge(df1, df2, on=rightKey, suffixes=['_x', '_y'], how=type)
    else:
        df = dd.merge(df1, df2, left_index=True, right_index=True, suffixes=['_x', '_y'], how=type)

    return df
```
